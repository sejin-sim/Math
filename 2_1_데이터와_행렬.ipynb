## 1. 데이터의 유형
### 1) 스칼라(scalar)
1. 스칼라 : 숫자 하나만으로 이루어 데이터
2. ex) 붓꽃 한송이의 꽃잎 길이 
> $ x \in R $

<br/>

### 2) 벡터(vector)
1. 벡터 : 여러 개의 숫자가 특정한 순서대로 모여 있는 것 (데이터 묶음)
2. ex) 붓꽃의 크기 벡터는 4개의 데이터 $  (x_1,x_2,x_3,x_4) $ 가 묶여있는 것
> $ \begin{bmatrix} x_1
\\ x_2
\\ x_3
\\ x_4
\end{bmatrix} $
> or
> $ x \in R^4 $
3. 하나의 벡터를 이루는 데이터 수가 $n$개 이면 벡터를 n-차원 벡터라고 한다.
> $ \begin{bmatrix} x_1
\\ x_2
\\ \cdot
\\ \cdot
\\ x_N
\end{bmatrix} $
> or
> $ x \in R^N $

<br/>

### 3) 특징 벡터(feature vector)
1. 특징 벡터 : 데이터 벡터가 예측 문제에서 입력 데이터로 사용되면
2. ex) 꽃받침 길이가 5.1cm, 꽃받침 폭이 3.5cm, 꽃잎 길이가 1.4cm, 꽃잎 폭이 0.2인 붓꽃 표본 한 송이(데이터 레코드)를 $x_1$라고 한다면?
> $ x_1 = \begin{bmatrix} 5.1
\\ 3.5
\\ 1.4
\\ 0.2
\end{bmatrix} $ $\;$ 
> $ x_2 = \begin{bmatrix} 4.9
\\ 3.0
\\ 1.4
\\ 0.2
\end{bmatrix} $
* 이러한 붓꽃 크기 벡터를 이용하여 붓꽃 종을 결정하는 예측 문제를 풀고 있다면 붓꽃 크기 벡터는 특징 벡터

```
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1))  # 벡터화
    
plt.figure(figsize=(8, 3))
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(v[i], aspect=0.4,
               interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("벡터 {}".format(i + 1))
plt.suptitle("벡터화된 이미지", y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

## 2. 전치 연산(transpose)
1) 전치 연산 : 행렬에서 가장 기본이 되는 연산으로 행렬의 행과 열을 바꾸는 연산
> $ x \;\; \rightarrow \;\; x^T $ 
> or
> $ x \;\; \rightarrow \;\; x' $
> * 전치 연산은 벡터나 행렬에 $T$라는 위첨자(superscript)를 붙여서 표기한다. 책에 따라서는 프라임(prime)기호 $'$를 붙이는 경우도 있다. 
2. ex) 앞에서 보인 $6\times 4$ 차원의 행렬을 전치 연산하면 $4\times 6$ 차원의 행렬이 된다.
> $ X = 
\begin{bmatrix}
\boxed{\begin{matrix} x_{1, 1} & x_{1, 2} & x_{1, 3} & x_{1, 4}\end{matrix}}  \\
\begin{matrix} x_{2, 1} & x_{2, 2} & x_{2, 3} & x_{2, 4}\end{matrix} \\
\begin{matrix} x_{3, 1} & x_{3, 2} & x_{3, 3} & x_{3, 4}\end{matrix} \\
\begin{matrix} x_{4, 1} & x_{4, 2} & x_{4, 3} & x_{4, 4}\end{matrix} \\
\begin{matrix} x_{5, 1} & x_{5, 2} & x_{5, 3} & x_{5, 4}\end{matrix} \\
\begin{matrix} x_{6, 1} & x_{6, 2} & x_{6, 3} & x_{6, 4}\end{matrix} \\
\end{bmatrix}
\;\; \rightarrow \;\; X^T = \begin{bmatrix}
\boxed{\begin{matrix} x_{1, 1} \\ x_{1, 2} \\ x_{1, 3} \\ x_{1, 4}\end{matrix}} &
\begin{matrix} x_{2, 1} \\ x_{2, 2} \\ x_{2, 3} \\ x_{2, 4}\end{matrix} &
\begin{matrix} x_{3, 1} \\ x_{3, 2} \\ x_{3, 3} \\ x_{3, 4}\end{matrix} &
\begin{matrix} x_{4, 1} \\ x_{4, 2} \\ x_{4, 3} \\ x_{4, 4}\end{matrix} &
\begin{matrix} x_{5, 1} \\ x_{5, 2} \\ x_{5, 3} \\ x_{5, 4}\end{matrix} &
\begin{matrix} x_{6, 1} \\ x_{6, 2} \\ x_{6, 3} \\ x_{6, 4}\end{matrix} &
\end{bmatrix}
$
3.  행 벡터(row vector) : (열)벡터 $x$에 대해 전치 연산을 적용하여 만든 $x^T$
> $ x = 
\begin{bmatrix}
x_{1} \\
x_{2} \\
\vdots \\
x_{N} \\
\end{bmatrix}
\; \rightarrow \;
x^T = 
\begin{bmatrix}
x_{1} & x_{2} & \cdots & x_{N}
\end{bmatrix}$

- NumPy에서는 `ndarray` 객체의 `T`라는 속성을 이용하여 전치 행렬을 구한다. 이때 `T`는 메서드(method)가 아닌 속성(attribute)이므로 소괄호 ``()``를 붙여서 호출하면 안 된다.

## 3. 행렬의 행 표기법과 열 표기법
1. 전치 연산과 행 벡터, 열 벡터를 이용하면 다음처럼 행렬을 복수의 열 벡터 $c_i$, 또는 복수의 행 벡터 $r_j^T$ 을 합친(concatenated) 형태로 표기할 수도 있다.
> $ X =
\begin{bmatrix}
{c_1, c_2, \cdots , c_M}
\end{bmatrix}
=
\begin{bmatrix}
{r_1^T}  \\
{r_2^T}  \\ 
{\vdots} \\ 
{r_N^T}  \\  
\end{bmatrix}
$

2. 위 식에서 행렬과 벡터의 크기는 다음과 같다.
> $ {X} \in \mathbf{R}^{N\times M}$   
> $ {c_i} \in \mathbf{R}^{N \times 1} \; (i=1,\cdots,M) $   
> $ {r_j^T} \in \mathbf{R}^{1 \times M} \; (j=1,\cdots,N) $

3. 벡터의 모양을 직사각형으로 표시하면 다음과 같다.
> $ X = 
\begin{bmatrix}
\boxed{\begin{matrix} \phantom{\LARGE\mathstrut} \\ {c_1} \\ \phantom{\LARGE\mathstrut} \end{matrix}} & 
\boxed{\begin{matrix} \phantom{\LARGE\mathstrut} \\ {c_2} \\ \phantom{\LARGE\mathstrut} \end{matrix}} & 
\cdots &
\boxed{\begin{matrix} \phantom{\LARGE\mathstrut} \\ {c_M} \\ \phantom{\LARGE\mathstrut} \end{matrix}} 
\end{bmatrix}
=
\begin{bmatrix}
\boxed{\begin{matrix} \phantom{} & \phantom{} & {r_1^T} & \phantom{} & \phantom{} \end{matrix}} \\ 
\boxed{\begin{matrix} \phantom{} & \phantom{} & {r_2^T} & \phantom{} & \phantom{} \end{matrix}} \\ 
\vdots \\ 
\boxed{\begin{matrix} \phantom{} & \phantom{} & {r_N^T} & \phantom{} & \phantom{} \end{matrix}} \\ 
\end{bmatrix} 
$

4. 위의 식은 다음과 같은 명제를 수식으로 표현한 것이다.
> * "행렬 $X$의 각 열은 $c_1$, $c_2$, $\ldots$, $c_M$라고 이름 붙인다."    
> * "행렬 $X$는 열 벡터 $c_1$, $c_2$, $\ldots$, $c_M$으로 이루어져 있다."   
> * "행렬 $X$의 각 행은 $r_1^T$, $r_2^T$, $\ldots$, $r_N^T$라고 이름 붙인다." 
> * "행렬 $X$는 행 벡터 $r_1^T$, $r_2^T$, $\ldots$, $r_N^T$으로 이루어져 있다"      

5. 모든 벡터는 기본적으로 열벡터이므로 $r_i$를 전치 연산하여 $r_i^T$라고 행을 표현한 점에 주의한다.
> ex) 행렬
$ 
X=
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
$
은   
> <br> 열 벡터 $c_1=
\begin{bmatrix}
1 \\
4 
\end{bmatrix},
\;
c_2=
\begin{bmatrix}
2 \\
5 
\end{bmatrix},
\;
c_3=
\begin{bmatrix}
3 \\
6 
\end{bmatrix} $이 합쳐진 것    
> <br> 또는 행 벡터 $ r_1=
\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}^T,
\;
r_2=
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}^T$이 합쳐진 것

## 4. 특수한 벡터와 행렬
### 1) 영벡터 
1. 영벡터(zeros-vector) : 모든 원소가 0인 $N$차원 벡터
> $ \mathbf{0}_N = \mathbf{0} = 0 =
\begin{bmatrix}
0 \\
0 \\
\vdots \\
0 \\
\end{bmatrix}
$   
> $
0 \in \mathbf{R}^{N \times 1} $
2. 문맥으로 벡터의 크기를 알 수 있을 때는 크기를 나타내는 아래 첨자 $N$을 생략할 수 있다.


```
np.zeros((3, 1))
```

### 2) 일벡터
1. 일벡터(ones-vector) : 모든 원소가 1인 $N$차원 벡터
> $
\mathbf{1}_N = \mathbf{1}  = 1 = 
\begin{bmatrix}
1 \\
1 \\
\vdots \\
1 \\
\end{bmatrix}
$   
> $ 1 \in \mathbf{R}^{N \times 1} $
2. 문맥으로 벡터의 크기를 알 수 있을 때는 크기를 나타내는 아래 첨자 $N$을 생략할 수 있다.

```
np.ones((3, 1))
```

### 4) 대각행렬
1. 대각(diagonal) : 행렬에서 행과 열이 같은 위치에 있는 것들
2. 비대각(off-diagonal) : 대각 위치에 있지 않은 것들
3. 대각행렬(diagonal matrix) : 모든 비대각 요소가 0인 행렬
>$ 
D = 
\begin{bmatrix}
d_{1} & 0 & \cdots & 0 \\
0 & d_{2} & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & d_{N} \\
\end{bmatrix}
$   
>$ D \in \mathbf{R}^{N \times N} $
4. 대각행렬이 되려면 비대각성분이 0이기만 하면 된다.
> $ 
D = 
\begin{bmatrix}
d_{1} & 0 & \cdots & 0 \\
0 & d_{2} & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & d_{M} \\
0 & 0 & \cdots & 0 \\
0 & 0 & \cdots & 0 \\
0 & 0 & \cdots & 0 \\
\end{bmatrix}
$   
> $ D \in \mathbf{R}^{N \times M} $

```
np.diag([1, 2, 3])
```

### 5) 항등행렬
1. 항등행렬(identity matrix) : 대각행렬 중에서도 모든 대각성분의 값이 1인 대각행렬
> $I = 
\begin{bmatrix}
1 & 0 & \cdots & 0 \\
0 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & 1 \\
\end{bmatrix}
$   
> $ 
I \in \mathbf{R}^{N \times N}
$

```
np.identity(3)
```

```
np.eye(4)
```

### 6) 대칭행렬
1. 대칭행렬(symmetric matrix) :전치연산을 통해서 얻은 전치행렬과 원래의 행렬이 같을때. 정방행렬만 대칭행렬이 될 수 있다.
> $ S^{T} = S $    
> $ S \in \mathbf{R}^{N \times N} $

2. 위의 벡터와 행렬을 NumPy로 나타내 본다.

```
# 영벡터, 일벡터, 대각행렬, 항등행렬, 정방행렬 = 대칭행렬
print("<3차원 영벡터>")
print(np.zeros((3, 1)))
print("\n<3차원 일벡터>")
print(np.ones((3, 1)))
print("\n<4x4 대각행렬>")
print(np.diag([1, 2, 3, 4]))
print("\n<4x4 항등행렬>")
print(np.identity(4))
print("- 지금까지의 행렬은 정방행렬이자 대칭행렬")
```
