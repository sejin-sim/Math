# 1. 기댓값과 확률변수의 변환
### 1) 확률변수의 기댓값
1. 기댓값(expectation) : 확률변수의 이론적 평균. 확률(또는 확률밀도)가 모여 있는 곳의 위치
2. 이산확률변수(X)의 기댓값 = 표본공간의 원소 $x_i$의 가중평균
> $ \mu_X = \text{E}[X] = \sum_{x_i \in \Omega} x_ip(x_i) $   
> ㄴ 가중치 = 확률질량함수 $p(x_i)$
2.-ex. 공정한 주사위에서 나올 수 있는 숫자를 대표하는 이산확률변수 $X$는 나올 수 있는 값이 1, 2, 3, 4, 5, 6
> $ \begin{aligned}
\mu_X 
&= 1 \cdot p(1) + 2 \cdot p(2) + 3 \cdot p(3) + 4 \cdot p(4) + 5 \cdot p(5) + 6 \cdot p(6) \\
&= 1 \cdot \dfrac{1}{6} + 2 \cdot \dfrac{1}{6} + 3 \cdot \dfrac{1}{6} + 4 \cdot \dfrac{1}{6} + 5 \cdot \dfrac{1}{6} + 6 \cdot \dfrac{1}{6} \\
&= \dfrac{7}{2} = 3.5
\end{aligned} $
3. 연속확률변수(X)의 기댓값 = 표본 $x$를 적분한 값
> $ \mu_X = \text{E}[X] = \int_{-\infty}^{\infty} x p(x) dx $


<br/>

### 2) 기댓값의 성질
1. 확률변수가 아닌 상수 $c$에 대해 
> $ \text{E}[c] = c $
2. 선형성
> $ \text{E}[cX] = c \text{E}[X] $   
> $  \text{E}[X + Y] = \text{E}[X] + \text{E}[Y] $    
> $ \text{E}[c_1X + c_2Y] = c_1\text{E}[X] + c_2\text{E}[Y] $ 
3. 확률변수 곱이 아니면 뭐든지 분리가능



3-ex. 회전하는 원반을 이용하여 복권 번호를 결정하는 문제.
> * $xp(x)$ = 연속확률변수(X)의 기댓값 = 확률밀도함수 $p(x)$ x $x$    
> * $ E[X] = xp(x)\text{의 면적} = \dfrac{1}{2} \cdot 360 \cdot 1 = 180 $

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib as mpl
import matplotlib.font_manager as fm
import warnings

warnings.filterwarnings(action='ignore') 

%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

```
x = np.linspace(-100, 500, 1000)
p = np.zeros_like(x)
p[(0 < x) & (x <= 360)] = 1 / 360
xp = x * p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title("$p(x)$ = 확률밀도함수")
plt.xlabel("$x$ (도)")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title("$xp(x)$ = 기댓값")
plt.xlabel("$x$ (도)")

plt.show()
```

```
# if 0도에서 180도 사이에 화살이 2배 더 잘 박히도록 원반이 조작되었다면,

x = np.linspace(-100, 500, 1000)
p = np.zeros_like(x)
p[(0 < x) & (x <= 180)] = 2 / (3 * 360)
p[(180 < x) & (x <= 360)] = 1 / (3 * 360)
xp = x * p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title("$p(x)$ = 확률밀도함수")
plt.xlabel("$x$ (도)")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title("$xp(x)$ = 기댓값")
plt.xlabel("$x$ (도)")\

plt.show()  
```

### 5) 표본평균 확률변수
1. 표본평균 확률변수 : 확률변수로부터 $N$개의 표본을 만들어 이 표본집합의 표본평균을 계산하여 나온 표본평균 값인 확률변수
2. 확률변수 $X$에서 나온 표본으로 만들어진 표본평균 확률변수 $\bar{X}$
> $ \bar{X} = \dfrac{1}{N}\sum_{i=1}^{N} X_i $

### 6) 기댓값과 표본평균의 관계
1. 표본평균 : 확률변수의 기댓값 근처의 값
> $ \text{E}[ \bar{X} ] = \text{E}[X] $
2. 증명
> $ \begin{aligned}
\text{E}[\bar{X}] 
&= \text{E}\left[\dfrac{1}{N}\sum_{i=1}^{N}X_i \right]  \\
&= \dfrac{1}{N}\sum_{i=1}^{N}\text{E}[X_i]  \\
&= \dfrac{1}{N}\sum_{i=1}^{N}\text{E}[X] \\
&= \dfrac{1}{N} N \text{E}[X] \\ 
&= \text{E}[X]  \\
\end{aligned} $
