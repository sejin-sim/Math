### 1) 선형종속과 선형독립

1. 선형종속(linearly dependent) : 벡터 집합 $x_1, x_2, \ldots, x_N$을 이루는 벡터의 선형조합이 영벡터가 되도록 하는 스칼라 계수 $c_1, c_2, \ldots, c_N$이 존재

- 스칼라와 벡터를 잘 조합하면 0을 만들 수 있는 특수한 경우
$$ 
\begin{align}
c_1 x_1 + c_2 x_2 + \cdots + c_N x_N = 0  
\end{align}
$$

2. 선형독립(linearly independent) : 벡터들의 선형조합이 0이 되면서 모두 0은 아닌 계수들이 존재하지 않는다.

$$ 
\begin{align}
c_1 x_1 + \cdots + c_N x_N = 0  \;\; \rightarrow \;\; c_1 = \cdots = c_N = 0  
\end{align}
$$ 
> 왼쪽에서 오른쪽 방향 화살표의 의미는 벡터들의 선형조합이 0이면 반드시 계수들이 모두 0이라는 뜻

</br>

- ex) 선형 독립의 예

$$ 
\begin{align}
x_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \;\;
x_2 = \begin{bmatrix} 3 \\ 3 \end{bmatrix}
\end{align}
$$
 > 어떤 계수를 사용해도 $c_1x_1 + c_2x_2 = 0$을 만들 수 없다.

- ex) 선형 종속의 예

$$ 
x_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \;\;
x_2 = \begin{bmatrix} 3 \\ 3 \end{bmatrix}, \;\;
x_3 = \begin{bmatrix} 5 \\ 7 \end{bmatrix}
$$

$$ 
\begin{align}
2x_1  + x_2 - x_3 = 0
\end{align}
$$

```
x1 = np.array([1, 2])
x2 = np.array([3, 3])
x3 = np.array([10, 14])
2 * x1 + x2 - 0.5 * x3
```

#### 연습 문제 3.2.2

서로 직교하는 영벡터가 아닌 $N$개의 벡터 $v_1, v_2, \cdots, v_N$은 선형독립임을 증명하라.

$$
\begin{align}
v_i^T v_j = 0 \;\;\; (i \neq j)
\tag{3.2.10}
\end{align}
$$

> 답 : 선형종속을 만드는 계수 존재시,

 $c_{1}v_{1} + \cdots + c_{N}v_{N} =0 $ 성립하고 $v_{1}$에 내적하면,

 $v_{1}(c_{1}v_{1} + \cdots + c_{N}v_{N}) =0 $

 벡터들은 서로 직교함으로 $v^{T}_{i}v_{j} = 0$
$v_{1}(c_{1}v_{1} + \cdots + c_{N}v_{N}) = c_{i}v^{T}_{i}v_{i} = c_{i}\left \|v_{i}\right \|^{2} =0 $

 벡터들은 0 벡터가 아니므로,

 $c_{i} = 0 $이고, 모든 $i=1, \cdots , N$ 대해 성립하므로 

 선형종속을 만드는 모두 0이 아닌 계수가 존재하지 않는다. 

 따라서 벡터 $v_1, v_2, \cdots, v_N$는 선형 독립이다.

### 2) 선형독립과 선형연립방정식

$$ \begin{align} c_1x_1 + c_2x_2 + \cdots + c_Nx_N & = \begin{bmatrix} x_1 & x_2 & \cdots & x_N \end{bmatrix} \begin{bmatrix} c_1 \\ c_2 \\ \vdots \\ c_N \end{bmatrix} \\ & = Xc \ \end{align} $$

- 선형 독립을 논리기호로

$$ Xc = 0 \;\; \rightarrow \;\; c=0  $$

### 4) 랭크

- 서로 독립인 벡터의 최대 개수 = $\text{rank}A$ = 열랭크 = 행랭크

- 랭크는 행과 열의 갯수보다 커질 수 없다.
$$ 
\begin{align}
\text{rank} A \leq \min(M, N)
\end{align}
$$

$$
\begin{align}
A \in \mathbf{R}^{N \times M}
\end{align}
$$

- ex) 두 열벡터는 선형 독립이다 = 랭크 2
$$ X_1 = \begin{bmatrix} 1 & 3 \\ 2 & 3 \end{bmatrix} $$

- ex) 다음 행렬은 행 < 열 = 선형 종속이다. 단, 앞의 두 열 벡터는 서로 독립이므로 = 랭크 2
$$ 
\begin{align}
X_2 = \begin{bmatrix} 1 & 3 & 5 \\ 2 & 3 & 7 \end{bmatrix} 
\end{align}
$$


```
# 랭크 계산 

X1 = np.array([[1, 3], [2, 3]])
print(np.linalg.matrix_rank(X1))

X2 = np.array([[1, 3, 5], [2, 3, 7]])
print(np.linalg.matrix_rank(X2))
```

> $$ \text{rank}A = \min(M, N)  $$

- 항상 풀랭크인 경우 : 선형독립인 벡터들을 행 또는 열로 가지는 행렬을 만들 때

### 6) 로우-랭크 행렬

1. 랭크-1 행렬(rank-1 matrix) : $N$차원 벡터 $x$ 하나를 이용하여 만들어지는 다음과 같은 행렬
> $$xx^T = x \begin{bmatrix} x_1 & x_2 & \cdots & x_n \end{bmatrix} = \begin{bmatrix} x_1x & x_2x & \cdots & x_nx \end{bmatrix} $$

2. 랭크-2 행렬(rank-2 matrix) : 선형독립인 두 개의 $N$차원 벡터 $x_1, x_2$를 이용하여 만든 다음과 같은 행렬
> $$
\begin{align}
\begin{bmatrix} x_1 & x_2 \end{bmatrix}
\begin{bmatrix} x_1^T \\ x_2^T \end{bmatrix}
= x_1x_1^T + x_2x_2^T
\end{align}
$$

3. 랭크-M 행렬(rank-M matrix) : $M$개의 $N$차원 벡터 $x_1, x_2, \cdots, x_M$을 이용
> 
$$
\begin{align}
\begin{bmatrix} x_1 & x_2 & \cdots & x_M \end{bmatrix}
\begin{bmatrix} x_1^T \\ x_2^T \\ \vdots \\ x_M^T \end{bmatrix}
= x_1x_1^T + x_2x_2^T + \cdots + x_Mx_M^T
= \sum_{i=1}^M x_ix_i^T
\end{align}
$$

- 1~3 까지의 행렬 = 로우-랭크 행렬(low-rank matrix) : 특이분해(SVD)와 PCA에서 사용

### 7) 벡터공간과 기저벡터

1. 벡터공간(vector space) : 벡터 N개가 서로 선형독립이면 이 벡터들을 선형조합하여 만들어지는 모든 벡터의 집합 $V$
> $$ V = \{ c_1x_1 + \cdots + c_Nx_N \; \vert \; c_1, \ldots, c_N \in \mathbf{R} \} $$

2. 기저벡터(basis vector) :  벡터 공간을 만들 수 있는 선형 독립인 벡터 (선형종속은 불가능)

1. 기저벡터가 정규직교이면, 투영벡터는 각 기저벡터에 대한 내적값으로 표현된다.
> $x^{||v}= (x^Tv_1) v_1+(x^Tv_2) v_2+...+(x^Tv_n) v_n$

2. 투영벡터의 길이
> $||x^{||v}||^2 =\Sigma(x^Tv_i)^2$

3. 직교 벡터$x^{\perp V}$는 벡터 공간의 모든 벡터에 대해서 직교한다.
> 투영벡터 $x^{\Vert V}$은 벡터공간의 모든 벡터 중에서 가장 벡터 $x$와 가까운 벡터

### 11) 표준기저벡터

1. 표준기저벡터(standard basis vector) : 기저벡터 중에서도 원소 중 하나만 값이 1이고 다른 값은 0으로 이루어진 다음과 같은 기저벡터
> 
$$ 
\begin{align}
e_1 = \begin{bmatrix} 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}, \;\;
e_2 = \begin{bmatrix} 0 \\ 1 \\ \vdots \\ 0 \end{bmatrix}, \;\; 
\cdots,  \;\;
e_N = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 1 \end{bmatrix}
\end{align}
$$

2. 표준기저벡터를 열로 가지는 행렬은 항등행렬이 된다.
> $$
\begin{align}
\begin{bmatrix} e_1 & e_2 & \cdots & e_N \end{bmatrix} = I_N
\end{align}
$$

### 12) 좌표

1. 좌표(coordinate) : 기저 벡터를 선형 조합하여 그 벡터를 나타내기 위한 계수 벡터

2. 벡터$(x)$와 기저벡터$(e_1 e_2)$ 그리고 좌표$(x_e )$의 관계
> $$ \begin{align}
x = [ e_1 e_2 ] \begin{bmatrix} x_{e_1} \\ x_{e_2} \end{bmatrix} = [ e_1 e_2 ] \; x_e 
\end{align}
$$

3. 같은 벡터라도 다른 기저벡터를 사용하면 좌표가 달라진다.

4. ex) 다음 기저벡터 $\{g_1, g_2\}$를 사용 시, x의 좌표 4, 2

$$
\begin{align}
g_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \; g_2 = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
\end{align}
$$

$$ 
\begin{align}
x = 4 g_1 + 2 g_2  
= \begin{bmatrix} g_1 & g_2 \end{bmatrix} \begin{bmatrix} 4 \\ 2 \end{bmatrix} 
= \begin{bmatrix} g_1 & g_2 \end{bmatrix} x_g
\end{align}
$$


```
e1 = np.array([1, 0])
e2 = np.array([0, 1])
x = np.array([2, 2])

plt.annotate('', xy=2 * e1, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=2 * e2, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.title("표준 좌표계 기저벡터")

g1 = np.array([1, 0])
g2 = np.array([-1, 1])
x = np.array([2, 2])
plt.show()

plt.annotate('', xy=4 * g1, xytext=(0, 0), arrowprops=lightgreen)
plt.annotate('', xy=2 * g2, xytext=(0, 0), arrowprops=lightgreen)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$g_2$", fontdict={"size": 18})
plt.text(0.5, -0.5, "$g_1$", fontdict={"size": 18})

plt.xticks(np.arange(-10, 10))
plt.yticks(np.arange(-10, 10))
plt.xlim(-3, 7)
plt.ylim(-2, 5)
plt.title("g1, g2 기저벡터")
plt.show()
```

### 13) 변환 행렬

1. 예를 들어 기존의 기저벡터 $\{e_1, e_2\}$와 새로운 기저벡터  $\{g_1, g_2\}$ 간의 다음과 같은 관계 성립시


$$
\begin{align}
\begin{aligned}
g_1 &=  \dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2 \\
g_2 &= -\dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2
\end{aligned}
\end{align}
$$

2. $e_1$, $e_2$에 대한 $g_1$, $g_2$의 좌표벡터

$$
\begin{align}
g_{1e} = 
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}, \;\;
g_{2e} = 
\begin{bmatrix} 
-\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\end{align}
$$

3. 두 좌표벡터들을 합쳐서 행렬로 표시

$$
\begin{align}
\begin{bmatrix} g_1 & g_2 \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} \begin{bmatrix} g_{1e} & g_{2e} \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} A
\end{align}
$$


$$ 
\begin{align}
A =
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\end{align}
$$

```
e1 = np.array([1, 0])
e2 = np.array([0, 1])
a = np.array([2, 2])
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)

plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=red)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=red)

plt.text(-0.18, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})
plt.text(0.3, 0.5, "$g_1$", fontdict={"size": 18})
plt.text(-0.45, 0.2, "$g_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.2, 1.7)
plt.ylim(-0.5, 1.3)
plt.title("변환 행렬")
plt.show()
```

### 14) 좌표변환

1. 좌표변환(coordinate transform) : 새로운 기저벡터에 대해 좌표를 계산하는 것.

2. 계산 방법   ( $
\begin{bmatrix} g_1 & g_2 \end{bmatrix}
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} 
A
$)

$$ 
\begin{align}
x 
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} x_e 
= \begin{bmatrix} g_1 & g_2 \end{bmatrix} x_g 
= \begin{bmatrix} e_1 & e_2 \end{bmatrix} A x_g
\end{align}
$$


$$ 
\begin{align}
x_e = A x_g 
\end{align}
$$


$$ 
\begin{align}
x_g = A^{-1}x_e = Tx_e 
\end{align}
$$

- 이 때 $A$의 역행렬 $T=A^{-1}$을 변환행렬(transform matrix)이라고 한다.

3. ex) 벡터 $x$의 표준기저벡터에 대한 좌표가 다음과 같은 경우


$$
\begin{align}
x = 2 e_1 + 2 e_2= \begin{bmatrix}2 \\ 2\end{bmatrix} = x_e
\end{align}
$$


$$ 
\begin{align}
g_{1e} = \begin{bmatrix} \dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix} ,\;\;
g_{2e} = \begin{bmatrix} -\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix}
\end{align}
$$


$$ 
\begin{align}
x_g = A^{-1}x_e = 
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} &  \dfrac{1}{\sqrt{2}} 
\end{bmatrix}^{-1}
\begin{bmatrix}2 \\ 2\end{bmatrix}
=
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} \\
-\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\begin{bmatrix}2 \\ 2\end{bmatrix} 
= \begin{bmatrix}2\sqrt{2}\\0\end{bmatrix}
\end{align}
$$

4. 결론 : 새로운 좌표벡터는 원래의 좌표벡터에 변환행렬을 곱하여 구할 수 있다.

```
e1 = np.array([1, 0])
e2 = np.array([0, 1])
x = np.array([2, 2])
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)

plt.annotate('', xy=e1, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=e2, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=x, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=g1, xytext=(0, 0), arrowprops=red)
plt.annotate('', xy=g2, xytext=(0, 0), arrowprops=red)

plt.plot(0, 0, 'ro', ms=10)
plt.plot(x[0], x[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$x$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})
plt.text(0.2, 0.5, "$g_1$", fontdict={"size": 18})
plt.text(-0.6, 0.2, "$g_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.title("기저벡터 변경에 따른 좌표 변환 : (2, 2) → ($2\sqrt{2}$, 0)")
plt.show()
```

```
A = np.vstack([g1, g2]).T
print(A)

Ainv = np.linalg.inv(A)
print(Ainv)

Ainv.dot(x)
```

### 15) 이미지 변환

1. 새로운 기저벡터에 대한 좌표변환을 응용하면 이미지를 자유롭게 변환할 수도 있다. 

2. 파이썬 이미지에서는 다음과 같은 표준기저벡터를 사용하고 ($x_1$이 아래를 향하는 세로축, $x_2$가 오른쪽을 향하는 가로축) **원점이 왼쪽 상단의 점**이라는 점에 유의한다.


$$ 
\begin{align}
e_1 = \begin{bmatrix} 0 \\ -1 \end{bmatrix} ,\;\;
e_2 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
\end{align}
$$


```
import scipy as sp
import scipy.misc
import scipy.ndimage

f = sp.misc.face(gray=True)

e1 = np.array([0, 1]) # matplotlib의 이미지 좌표규약으로 부호변경 
e2 = np.array([1, 0])
E = np.vstack([e1, e2]).T
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)
A = np.vstack([g1, g2]).T
gc1 = E @ g1
gc2 = E @ g2

plt.subplot(121)
plt.imshow(f, cmap=mpl.cm.bone, alpha=0.9)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=green)
plt.annotate('$e_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=green)
plt.annotate('$e_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.annotate('', xy=500*gc1, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_1$', xy=500*gc1, xytext=500*gc1 + [50, -50])
plt.annotate('', xy=500*gc2, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_2$', xy=500*gc2, xytext=500*gc2 + [50, 0])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환전")

f1 = sp.ndimage.affine_transform(f, A)

plt.subplot(122)
plt.imshow(f1, cmap=mpl.cm.bone, alpha=0.8)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=red)
plt.annotate('$g_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=red)
plt.annotate('$g_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환후")
plt.show()
```
