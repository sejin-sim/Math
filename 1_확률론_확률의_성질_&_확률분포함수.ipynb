# 1. 확률의 성질
### 1) 공집합의 확률
1. 공집합인 사건의 확률은 0이다. $ P(\emptyset) = 0 $
>$ P(A \cup \emptyset) = P(A) = P(A) + P(\emptyset) $   
>$ \therefore P(\emptyset) = 0 $

<br/>

### 2) 여집합의 확률
1. 어떤 사건의 여집합인 사건의 확률은 (1 - 원래 사건의 확률)과 같다. $ P(A^C) = 1 - P(A) $
>if $ \;\ 0 \leq P(A) \leq 1,\;\ P(A \cup A^C) = P(\Omega) = 1 = P(A) + P(A^C)$   
>$ \therefore P(A^C) = 1 - P(A) $

<br/>

### 3) 포함-배제 원리
1. 두 사건의 합집합의 확률은 각 사건의 확률의 합에서 두 사건의 교집합의 확률을 뺀 것과 같다.
> $ P(A \cup B) = P(A) + P(B) - P(A \cap B) $
2. 증명
> $
\begin{aligned}
P(A \cup B) 
&= P(A \cup (B \cap A^C)) \\
&= P(A) + P(B \cap A^C) \\
&= P(A) + P(B \cap A^C) + P(A \cap B) - P(A \cap B) \\
&= P(A) + P((A^C \cap B) \cup (A \cap B)) - P(A \cap B) \\
&= P(A) + P(B) - P(A \cap B)
\end{aligned}
$

<br/>

### 4) 전체 확률의 법칙 (law of total probability)
1. 복수의 사건  $C_i$가 다음을 만족하는 사건들이라고 가정한다면, 
 * 서로 교집합이 없다. 이를 서로 배타적(mutually exclusive)이라고도 한다.
 * 모든 집합의 합집합이 전체집합(표본공간)이다. 이 경우 완전한(complete) 부분집합들이라고 한다.
 * if $ C_i \cap C_j = \emptyset \;\;\; (i \neq j) \;\;\; and \;\;\; C_1 \cup C_2 \cup \cdots  = \Omega,$
2. 사건 $A$의 확률은 사건 $A$와 사건 $C_i$가 동시에 발생할 사건들의 확률의 합과 같다.
> $P(A) = \sum_i P(A \cap C_i) $
3. ex) 생일이 1월인 남자의 집합이고 전체 확률의 법칙
> $P(남자) =  P(생일=1월인 남자) + P(생일=2월인 남자) + P(생일=이외 월의 남자) $
4. 증명
> $P(A) = P(A \cap C_1) + P(A \cap C_2) + \cdots = \sum_i P(A\cap C_i) = \sum_i P(A,C_i)  $

# 2. 확률분포함수

### 1) 단순사건과 확률질량함수
1. 단순사건 : 표본이 하나인 사건. 단순사건 끼리는 교집합 X → 다른 모든 사건의 확률값 계산 가능
> $
ex) \;\; P(\{\spadesuit\})=0.1, \;\; P(\{\heartsuit\})=0.2,  \;\; P(\{\diamondsuit\})=0.3,  \;\; P(\{\clubsuit\})=0.4$   
$\;\; \;\;\;\;\; P(\{\heartsuit, \diamondsuit\})=0.2+0.3=0.5$ 

</br>

2. 확률질량함수 $ p(a) $ : 각 단순사건에 대한 확률만 정의하는 함수, 이산 확률변수의 각 값에 대해 정의되는 함수
> $ p(a) = P(\{a\}) $   
> $ p(1) = P(\{1\}) = 0.2 $

3. 구간

> $ P(B) = P(\{ -2 < x \leq 1\}) + P(\{2 < x \leq 3\}) = P(-2, 1) + P(2, 3) $

### 2) 누적분포함수
1. 누적분포함수(cdf, $F(x)$) : 사건의 확률분포를 묘사하는 함수 ($x$ : 독립변수로 구간의 끝점)
> $ F(x) = P(S_x) = P(\{X < x\}) $
2. 구간 사건의 확률 with 콜모고로프의 공리
> $ A \cap B = \emptyset \;\;\; \rightarrow \;\;\; P(A \cup B) = P(A) + P(B) $   
> $ \therefore P(-\infty, b) = P(-\infty, a) + P(a, b) $
3. 누적함수로 표현
> $ F(b) = F(a) + P(a, b) $   
> $ \therefore P(a, b) = F(b) - F(a) $
4. 특징
 * $ F(-\infty) = 0 $   
 * $ F(+\infty) = 1 $   
 * $x > y \;\; \rightarrow \;\; F(x) \geq F(y)$

```
# 원반의 각도 문제에서 누적분포함수. 

t = np.linspace(-100, 500, 100)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
plt.plot(t, F)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.title("누적분포함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$F(x)$")
plt.show()
```

### 3) 확률밀도함수
* 누적분포함수는 음수 무한대가 시작점이어서 분포의 형상을 직관적으로 이해하기 힘듬
1. 확률밀도함수(probability density function, $p(x)$) : 누적분포함수를 미분하여 구한 도함수
> $x_1$ 근처에서 폭 $dx$를 가지는 구간의 확률   
> $ P(\{ x_1 < x \leq x_1 + dx \}) = F(x_1 + dx) - F(x_1) $
2. 단위 구간 길이당 배정된 확률값
> * $ \frac{P(\{ x_1 < x \leq x_1 + dx \})}{dx} $   
> * $ \begin{aligned}
\lim_{dx \rightarrow 0} \frac{P(\{ x_1 < x \leq x_1 + dx \})}{dx} 
&= \lim_{dx \rightarrow 0}  \frac{F(x_1 + dx) - F(x_1)}{dx} \\
&= x_1\text{에서} F(x)\text{의 기울기} \end{aligned}$

```
t = np.linspace(-100, 500, 1000)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
p = np.gradient(F, 600/1000)  # 수치미분
plt.plot(t, p)
plt.ylim(-0.0001, p.max()*1.1)
plt.xticks([0, 180, 360])
plt.title("확률밀도함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$p(x)$")
plt.show()
```
