### 1) 벡터의 기하학적 의미

N 차원 벡터 a는 N차원의 공간에서

1. 벡터 a의 값으로 표시되는 점(point) 또는

2. 원점과 벡터 a의 값으로 표시되는 점을 연결한 화살표(arrow)

> ex) 
$ a = \begin{bmatrix}a_{1} \\ a_{2} \end{bmatrix}
$

- 점(point) : 2차원 공간에서 $x$ 좌표가 $a_1$, $y$ 좌표가 $a_2$인 점으로 생각할 수도 있고 

- 화살표(arrow) : 원점에서 이 점을 가리키는 화살표로 생각할 수도 있다. 벡터를 화살표로 생각하는 경우에는 길이와 방향을 고정시킨 채 평행이동할 수 있다.

### 2) 벡터의 길이

1. 벡터 $a$의 길이 = 놈(norm) $\| a \|$으로 정의


$$ 
\| a \| = \sqrt{a^T a } = \sqrt{a_1^2 + \cdots + a_N^2}
$$

- 넘파이 linalg 서브 패키지의 `norm()` 명령으로 벡터의 길이를 계산할 수 있다. 
>  $a = [a_1 \;\; a_2]^T$의 길이는 $\sqrt{5} \approx 2.236$이다.

### 4) 단위 벡터

- 단위벡터(unit vector) : 길이가 1인 벡터

$$ 
a = \begin{bmatrix}1 \\ 0\end{bmatrix} ,\;\;
b = \begin{bmatrix}0 \\ 1\end{bmatrix} ,\;\;
c = \begin{bmatrix} \dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix}
$$

$$
\sqrt{1^{2} + 0^{2}} =  \sqrt{0^{2} + 1^{2}} =  \sqrt{{\dfrac{1}{\sqrt{2}}}^{2} + {\dfrac{1}{\sqrt{2}}}^{2}} = 1
$$

- 영벡터가 아닌 임의의 벡터 $x$에 대해 벡터 $x$와 같은 방향을 가리키는 단위벡터

$$

\dfrac{x}{\| x \|}

$$


```
a = np.array([1, 0])
b = np.array([0, 1])
c = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
np.linalg.norm(a), np.linalg.norm(b), np.linalg.norm(c)
```

### 6) 벡터의 선형조합

$$

```
x1 = np.array([1, 2])
x2 = np.array([2, 1])
x3 = 0.5 * x1 + x2
plt.annotate('', xy=0.5*x1, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x2, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=x3, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(x1[0], x1[1], 'ro', ms=10)
plt.plot(x2[0], x2[1], 'ro', ms=10)
plt.plot(x3[0], x3[1], 'ro', ms=10)
plt.plot([x1[0], 0], [x1[1], 0], 'k--')
plt.text(0.6, 2.0, "$x_1$")
plt.text(-0.5, 0.5, "$0.5x_1$")
plt.text(1.15, 0.25, "$x_2$")
plt.text(2.5, 1.6, "$0.5x_1 + x_2$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

### 7) 벡터의 차

1. 벡터의 차 $a - b = c$는 벡터 $b$가 가리키는 점으로부터 벡터 $a$가 가리키는 점을 연결하는 벡터
> 벡터 $b$와 벡터 $a-b$를 연결하면 벡터 $a$가 되어야 하기 때문이다.

$$
a - b = c
$$

$$
b + c = b + (a-b) = a
$$

```
a = np.array([1, 2])
b = np.array([2, 1])
c = a - b
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=a, xytext=b, arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 0.25, "$b$")
plt.text(1.55, 1.65, "$a-b=c$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-0.8, 2.8)
plt.ylim(-0.8, 2.8)
plt.show()
```

### 8) Word2Vec

1. word2vec : 단어(word)를 공간에서 점 또는 벡터(vector)로 표현할 수 있다. 

2. word2vec으로 만들어진 벡터는 단어의 의미에 따라 다음처럼 평행사변형 관계를 가질 수 있다. 

> 일본 = 도쿄 + 한국 - 서울


> `한국 - 서울`은 `서울`에서 `한국`으로 향하는 벡터다. 즉 의미론적으로 **수도 이름을 나라 이름으로 바꾸는 행위(action)**에 비유할 수 있다. 

> 이러한 행위를 `도쿄`에 대해서 적용한 결과가 `도쿄 + (한국 - 서울)`이다. word2vec 학습 결과에서 이렇게 계산한 위치에 가장 가까이 있는 단어를 찾으면 `도쿄`가 나온다.

```
a = np.array([2, 2])
b = np.array([3, 4])
c = np.array([4, 1])
d = a + (c - a)
e = b + (c - a)
plt.annotate('', xy=b, xytext=a, arrowprops=black)
plt.annotate('', xy=e, xytext=d, arrowprops=black)
plt.annotate('', xy=c, xytext=[0, 0], arrowprops=gray)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.text(1.6, 1.5, "서울")
plt.text(2.5, 4.3, "한국")
plt.text(3.5, 0.5, "도쿄")
plt.text(4.9, 3.2, "일본")
plt.xticks(np.arange(-2, 7))
plt.yticks(np.arange(-1, 6))
plt.xlim(-1.4, 6.4)
plt.ylim(-0.6, 5.8)
plt.show()
```

### 8) 유클리드 거리

1. 유클리드 거리(Euclidean distance) : 두 벡터가 가리키는 점 사이의 거리. 벡터의 차의 길이로 구할 수 있다.

2. 벡터의 놈의 정의와 벡터의 차의 정의에서 유클리드 거리는 다음처럼 구한다.

$$ 
\begin{aligned} 
\| a - b \|
&= \sqrt{\sum_{i=1} (a_i - b_i)^2} \\
&= \sqrt{\sum_{i=1} ( a_i^2 - 2 a_i b_i + b_i^2 )} \\
&= \sqrt{\sum_{i=1} a_i^2 + \sum_{i=1} b_i^2 - 2 \sum_{i=1} a_i b_i} \\
&= \sqrt{\| a \|^2 + \| b \|^2  - 2 a^Tb }
\end{aligned}
$$

<br/>

$$
\| a - b \|^2 = \| a \|^2 + \| b \|^2 - 2 a^T b
$$

### 9) 벡터의 내적과 삼각함수

1. 두 벡터의 내적 : 벡터의 길이 $\|a\|$, $\|b\|$ 와 두 벡터 사이의 각도 $\theta$의 코사인 함수값으로 계산할 수도 있다.

$$ 
a^Tb = \|a\|\|b\| \cos\theta
$$

2. $\cos\theta$ 코사인(cosine) : 사인(sine)이라고 하는 함수와 함께 정의할 수 있다. 

3. 삼각함수 = 사인 + 코사인

> $ \sin\theta = \dfrac{a}{h} = $ 높이/빗변
>> $\sin 0^{\circ} = 0, \;\sin 90^{\circ} = 1 $

> $ \cos\theta = \dfrac{b}{h} = $ 밑변/빗변   
>> $ \cos 0^{\circ} = 1 , \; \cos 90^{\circ} = 0 $


```
plt.plot([0, 1], [0, 2], 'k-', lw=3)
plt.plot([0, 1], [0, 0], 'k-', lw=3)
plt.plot([1, 1], [0, 2], 'k-', lw=3)
plt.text(0.05, 1, "빗변 h")
plt.text(0.35, -0.2, "밑변 b")
plt.text(1.05, 1, "높이 a")
plt.text(0.12, 0.06, r"$\theta$")
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.1, 2.1)
plt.ylim(-0.5, 2.3)
plt.show()
```

```
x = np.linspace(0, np.pi/2, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r--', lw=3,  label=r"$\sin\theta$")
plt.plot(x, y2, 'b-', lw=3, label=r"$\cos\theta$")
plt.legend()
plt.xticks([0, np.pi/4, np.pi/2], [r'$0^{\circ}$', r'$45^{\circ}$', r'$90^{\circ}$'])
plt.xlabel(r"$\theta$")
plt.title(r"$\sin\theta$와 $\cos\theta$의 그래프")
plt.show()
```

### 10) 직교

1. 직교(orthogonal) : 두 벡터 $a$와 $b$가 이루는 각이 90도이면 서로 직교한다고 한다.
> $\cos 90^{\circ} = 0$이므로  서로 직교인 두 벡터의 내적은 0이 된다.

$$ 
a^T b = b^T a = 0   \;\;\;\; \leftrightarrow \;\;\;\; a \perp b 
$$

### 11) 정규직교

1. 정규직교(orthonormal) : 만약 $N$개의 단위벡터($v_1, v_2, \cdots, v_N$)가 서로 직교할 때

$$ 
\|v_i\| = 1 \;\;\leftrightarrow \;\; v_i^Tv_i = 1 
$$

$$ 
v_i^T v_j = 0 \;\;(i \neq j) 
$$

### 14) 투영성분과 직교성분

1. 투영성분(projection) : 벡터 $a$를 다른 벡터 $b$에 직교하는 성분과 벡터 $b$에 평행한 성분으로로 분해할 수 있는데, 평행한 성분을 벡터 $b$에 대한 투영성분이라 한다.


$$ 
a^{\Vert b} 
$$


2. 직교성분(rejection) : 벡터 $b$에 직교하는 성분을 벡터 $b$에 대한 직교 성분이라 한다.

$$ 
a^{\perp b} 
$$

3. 투영성분의 길이

$$ 
\| a^{\Vert b} \| 
= \|a\|\cos\theta 
= \dfrac{\|a\|\|b\|\cos\theta}{\|b\|}  
= \dfrac{a^Tb}{\|b\|} 
= \dfrac{b^Ta}{\|b\|}
= a^T\dfrac{b}{\|b\|} 
$$

4. if 벡터 $b$ 자체가 이미 단위벡터이면, 단위벡터에 대한 투영길이 = 내적

$$ 
\| a^{\Vert b} \| = a^Tb 
$$

5. 투영성분 성분 벡터는 투영성분 길이와 벡터 $b$ 방향의 단위벡터의 곱이다.

$$ 
a^{\Vert b} = \dfrac{a^Tb}{\|b\|} \dfrac{b}{\|b\|}= \dfrac{a^Tb}{\|b\|^2}b  
$$

6. 직교성분 벡터는 원래의 벡터에서 투영성분 성분 벡터를 뺀 나머지다.

$$ 
a^{\perp b} = a - a^{\Vert b} 
$$



```
a = np.array([1, 2])
b = np.array([2, 0])
a2 = (a @ b) / np.linalg.norm(b) * np.array([1, 0])
a1 = a - a2
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=a2, xytext=(0, 0), arrowprops=blue)
plt.annotate('', xy=a1, xytext=(0, 0), arrowprops=blue)
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=red)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.text(0.35, 1.15, "$a$")
plt.text(1.55, 0.15, "$b$")
plt.text(-0.5, 1.05, "$a^{\perp b}$")
plt.text(0.50, 0.15, "$a^{\Vert b}$")
plt.xticks(np.arange(-10, 10))
plt.yticks(np.arange(-10, 10))
plt.xlim(-1.2, 4.1)
plt.ylim(-0.5, 3.2)
plt.show()
```

> (2) 벡터 $w$가 가리키는 점을 지남

$$
w^T(x - w) = 0

$$
w^T(x - w) = w^Tx - w^Tw = w^Tx - \| w \|^2
$$

$$
w^Tx - \| w \|^2 = 0
$$

4. ex) 예를 들어

$$
\begin{bmatrix}1 & 2\end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \end{bmatrix} - 5 = x_1 + 2x_2 - 5 = 0
$$

$$
x_1 + 2x_2 = 5 
$$

> 이 방정식은 벡터 $w$가 가리키는 점 $(1, 2)$를 지나면서 벡터 $w$에 수직인 직선을 뜻한다. 이 직선과 원점 사이의 거리는 $\|w\|=\sqrt{5}$이다.

5. 이번에는 벡터 $w$가 가리키는 점을 지나야 한다는 조건을 없애고, 단순히 벡터 $w$에 수직인 직선 $x$의 방정식을 구해보자.

- 이때는 직선이 $w$가 아니라 $w$와 방향이 같고 길이가 다른 벡터 $w'=cw$을 지날 것이다.

w'^Tx - \| w' \|^2 =  cw^Tx - c^2 \| w \|^2 = 0  
$$

w^Tx - c \| w \|^2 = 0

### 16) 직선과 점의 거리

1. 직선 $w^Tx - \|w\|^2 = 0$ 과 이 직선 위에 있지 않은 점 $x'$ 사이의 거리($d$) : $ x'^{\Vert w} $의 길이와 ${\|w\|}$의 차이

- 벡터 $x'$의 투영성분 $x'^{\Vert w}$의 길이

$$
\|x'^{\Vert w}\| = \dfrac{w^Tx'}{\|w\|} 
$$

$$
d = \left|  \|x'^{\Vert w}\| - \|w\| \right| = 
\left| \dfrac{w^Tx'}{\|w\|} - \|w\| \right| =
\dfrac{\left|w^Tx' - \|w\|^2 \right|}{\|w\|}
$$

2. 직선의 방정식이 $w^Tx - w_0 = 0$인 경우 직선과 점의 거리

$$
\dfrac{\left|w^Tx' - w_0 \right|}{\|w\|} 
$$

> - 이 공식은 나중에 분류 방법의 하나인 서포트 벡터 머신(SVM: Support Vector Machine)에서 사용된다.

$$
